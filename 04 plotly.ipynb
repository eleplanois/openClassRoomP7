```python
app_plt = app_train.sample(10000, random_state=1944)
```

```python
# creating trace1
trace1 =go.Scatter(
                    x = app_plt.EXT_SOURCE_1,
                    y = app_plt.CREDIT_TERM,
                    mode = "markers",
                    name = "EXT1",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= "EXT1")

# creating trace2
trace2 =go.Scatter(
                    x = app_plt.EXT_SOURCE_2,
                    y = app_plt.CREDIT_TERM,
                    mode = "markers",
                    name = "EXT2",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    text= "EXT2")

# creating trace3
trace3 =go.Scatter(
                    x = app_plt.EXT_SOURCE_3,
                    y = app_plt.CREDIT_TERM,
                    mode = "markers",
                    name = "EXT3",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    text= "EXT3")

data = [trace1, trace2, trace3]
layout = dict(title = 'Essai plotlib',
              xaxis= dict(title= 'Source EXT',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'CREDIT_TERM',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```
